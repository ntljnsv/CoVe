```python
import pandas as pd
from bert_score import score
```

```python
df = pd.read_csv("data.csv")
```

```python
original_sentences = df["Style 1"].tolist()
true_sentences = df["Style 2"].tolist()
paraphrased_sentences = df["Rephrased"].tolist()
```

Original sentence and rephrased sentence using CoVe

```python
P, R, F1 = score(paraphrased_sentences, original_sentences, lang="en", verbose=True)
print(f"Precision: {P.mean()}, Recall: {R.mean()}, F1: {F1.mean()}")
```

True rephrased sentence and rephrased sentence using CoVe

```python
P, R, F1 = score(paraphrased_sentences, true_sentences, lang="en", verbose=True)
print(f"Precision: {P.mean()}, Recall: {R.mean()}, F1: {F1.mean()}")
```
